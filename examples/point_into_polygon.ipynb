```python
from koala.pointsets import uniform
from koala.voronization import generate_lattice
from koala.example_graphs import higher_coordination_number_example
from koala.plotting import plot_edges, plot_vertex_indices
import numpy as np
from matplotlib import pyplot as plt

```

```python
from koala.graph_utils import vertices_to_polygon, make_dual

# generate the lattice
points = uniform(30)
lattice = generate_lattice(points)
lattice = make_dual(lattice)

# plot it
fig, ax = plt.subplots(figsize = (10,10))
plot_edges(lattice)
plot_vertex_indices(lattice)
```

```python
from random import sample

# pick a few vertices at random
i = sample(range(lattice.n_vertices), 10)

# turn these into polygons
modified = vertices_to_polygon(lattice, i)

# plot that scheisse
fig, ax = plt.subplots(1,2,figsize = (20,10))
ax[0].scatter(lattice.vertices.positions[i,0], lattice.vertices.positions[i,1], s = 1000)
plot_edges(lattice, ax = ax[0])
plot_edges(modified, ax = ax[1])
plot_vertex_indices(lattice, ax = ax[0])
plot_vertex_indices(modified,ax = ax[1])
plt.show()
```
